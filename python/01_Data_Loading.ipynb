**Objective:** This notebook initiates the data pipeline by loading raw campaign datasets for **Nykaa**, **Purplle**, and **Tira**, adding a brand tracker, concatenating them into a unified dataset, and performing initial structural audits.

### 1. Load brand campaign files and merge them
We check for workspace subdirectories dynamically so that the pipeline operates correctly whether it is run from the workspace root or from the `python/` directory itself.

```python
import os
import pandas as pd
import numpy as np

# Dynamic path resolution
data_dir = 'data'
if not os.path.exists(data_dir):
    data_dir = os.path.join('..', 'data')

# File paths
nykaa_path = os.path.join(data_dir, 'nykaa_campaign_data.csv')
purplle_path = os.path.join(data_dir, 'purplle_campaign_data.csv')
tira_path = os.path.join(data_dir, 'tira_campaign_data.csv')

# Load CSVs
nykaa = pd.read_csv(nykaa_path)
purplle = pd.read_csv(purplle_path)
tira = pd.read_csv(tira_path)

# Add brand identifiers
nykaa['Brand'] = 'Nykaa'
purplle['Brand'] = 'Purplle'
tira['Brand'] = 'Tira'

# Concatenate datasets
marketing = pd.concat([nykaa, purplle, tira], ignore_index=True)

# Save raw merged file
raw_output = os.path.join(data_dir, 'marketing_raw_merged.csv')
marketing.to_csv(raw_output, index=False)

print("Raw brand files loaded and merged successfully!")
print(f"Consolidated raw shape: {marketing.shape}")
print(f"Saved raw merged dataset to: {raw_output}")
```

### 2. Structural & Statistical Inspection
We inspect basic dataset parameters (`shape`, `columns`, `info()`, `describe(include='all')`) to verify column formats.

```python
# Display metadata
print("Columns:", marketing.columns.tolist())
print("-" * 80)
print("Dataset structural summary:")
marketing.info()
print("-" * 80)
print("Summary statistics (All Columns):")
marketing.describe(include='all')
```
